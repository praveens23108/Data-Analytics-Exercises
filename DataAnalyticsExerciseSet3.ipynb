```python
# amount of penguins on each island

df = sns.load_dataset('penguins')
penguins = df['island'].value_counts()
print(penguins)

# Torgersen island is not in balance with the other two islands.
```

```python
# code here

# visualization of the amount of penguins on each island

import matplotlib.pyplot as plt
sns.countplot(x='island', data=df)
plt.title('Number of Penguins on Each Island')
plt.xlabel('Island')
plt.ylabel('Count')
plt.show()
```

```python
# code here
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('penguins')

sns.boxplot(x='species', y='flipper_length_mm', hue='island', data=df)
plt.title('Flipper Length by Species and Island')
plt.show()
```

```python
sns.boxplot(x='species', y='bill_length_mm', hue='island', data=df)
plt.title('Bill Length by Species and Island')
plt.show()
```

```python
sns.boxplot(x='species', y='body_mass_g', hue='island', data=df)
plt.title('Body Mass by Species and Island')
plt.show()
```

**Adelie penguins have the shortest beak, Chinstrap penguins are in the middle, and Gentoo penguins are the largest (heaviest body, longest flippers) and exclusive to Biscoe Island.**
